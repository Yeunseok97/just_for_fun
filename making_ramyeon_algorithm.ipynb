```python
import time

boil_1 = """
    <
    >  >  >
    <  <  <
ㅣ           ㅣ
ㅣ           ㅣ-----------ㅣ
ㅣ           ㅣ-----------ㅣ
ㅣ___________ㅣ
"""

boil_3 = """
          <
    >  >  >
    <  <  <
ㅣ           ㅣ
ㅣ           ㅣ-----------ㅣ
ㅣ           ㅣ-----------ㅣ
ㅣ___________ㅣ
"""

boil_2 = """
       >
    >  >  >
    <  <  <
ㅣ           ㅣ
ㅣ           ㅣ-----------ㅣ
ㅣ           ㅣ-----------ㅣ
ㅣ___________ㅣ
"""

art1 = """
 ╭ ⁀ ⁀ ╮
( '👅'　　)
 ╰ ‿ ‿ ╯
　　　　　　　　╭ ⁀ ⁀ ╮
       　　　  ( '👅'　　)
　　　　　　　　╰ ‿ ‿ ╯
 ╭ ⁀ ⁀ ╮
( '👅'　　)
 ╰ ‿ ‿ ╯
"""
art2 = """
       >
    >  >  >
    <  <  <
|            |----------|⊂_ヽ
|　          |----------|    ＼＼ Λ＿Λ
|　　        |                 ＼( 'ㅅ' )
|------------|                   >　⌒ヽ
　　　                         / 　 へ＼
　　                          /　　/　＼＼
　　                         ﾚ　ノ　　  ヽ_つ
　  　                      /　/
　                        /　/|
  　                      (　(ヽ
  　                      |　|、＼
  　                      | 丿 ＼ ⌒)
  　                      | |　　) /
                        ノ )　　Lﾉ
"""
art3 = """
     ∧  _  ∧
    (。・ω・。)つ━☆・*。
   ⊂ ノ ・゜+.
       しーJ

      °。+ *´¨)
  .· ´¸.·*´¨) ¸.·*¨)
(¸.·´ (¸.·'*
"""



boil = [boil_1, boil_2, boil_3]


def print_boil(art_index):
    print(boil[art_index])


def ramyeon(물):
    if 물 == 'y' or 물 == 'Y':
        print("물을 끓이는 중입니다")
        for i in range(101):
            print_boil(i % 3)
            print("물 온도 : ",i, "도\n")
            time.sleep(0.1)
            if i == 100:
                print("물이 다 끓었습니다 ^^\n\n ")
                print(art1)
                print("\n\n")
                ramyeon_and_soup = input("라면과 스프를 넣어주세요 y/n: \n")
                if ramyeon_and_soup == 'y' or ramyeon_and_soup == 'Y':
                    print(art3)
                    print("라면과 스프를 넣었습니다!!")
                    for j in range(1, 6):
                        print_boil(j % 3)
                        print(j, "분 지났습니다.\n")
                        time.sleep(1)
                    print("٩(๑˃́ꇴ˂̀๑)و ")
                    cooked = input("면이 다 익었나요? (y/n): \n")
                    if cooked == 'y' or cooked == 'Y':
                          print(art2)
                          print()
                          print("라면 완성!!!")
                          break  # 면이 다 익었을 때 루프 종료
                    else:
                        print_boil(1)
                        print("라면을 계속 끓입니다.")
                        min = 1
                        while True:
                          additional_cook = input("면이 다 익었나요?? (y/n): ")
                          if additional_cook == 'y' or additional_cook == 'Y':
                            print(art2)
                            print()
                            print("라면 완성!!!")
                            break  # 루프 종료
                          else:
                            print_boil(min%3)
                            print("라면을 %d분 째 더 끓입니다. \n"%(min))
                            min += 1
                            continue
                else:
                    print("라면 끓이기를 중지합니다")
                    break  # 스프를 넣지 않을 때 루프 종료
    else:
        print("라면 끓이기를 중지합니다.")

물 = input("냄비에 물을 넣어주세요 y/n: ")
ramyeon(물)
```
